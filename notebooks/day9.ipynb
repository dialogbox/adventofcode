```python
f = open('../inputs/day9.txt', 'r')
lines = f.read().splitlines()
f.close()
```

```python
lines
```

```python
x = np.arange(len(lines[0]))
y = np.arange(len(lines))
```

```python
X2D,Y2D = np.meshgrid(x,y)

Y = Y2D.ravel()
X = X2D.ravel()
sizes = np.empty(len(X)); sizes.fill(5)
```

```python
colors=np.array([int(lines[j][i]) for (i,j) in list(zip(X,Y))])
```

```python
%matplotlib widget
from matplotlib import pyplot as plt
import numpy as np

fig = plt.figure()

fig.canvas.layout.width = '100%'
fig.canvas.layout.height = '900px'


plt.style.use('_mpl-gallery')
plt.scatter(X*100, Y*100, s=sizes, c=colors*100)
```
