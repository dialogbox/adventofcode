```python
from itertools import product


def is_conflict(c1, c2):
    (x1, y1, z1) = c1
    (x2, y2, z2) = c2
    
    points = list(product(x2, y2, z2))
    
    for (x, y, z) in points:
        if (x1[0] <= x and x <= x1[1] and
            y1[0] <= y and y <= y1[1] and
            z1[0] <= z and z <= z1[1]):
            return True
    return False

def is_fully_contained(c1, c2):
    (x1, y1, z1) = c1
    (x2, y2, z2) = c2
    
    points = list(product(x2, y2, z2))
    
    for (x, y, z) in points:
        if not (x1[0] <= x and x <= x1[1] and
            y1[0] <= y and y <= y1[1] and
            z1[0] <= z and z <= z1[1]):
            return False
    return True
    
def is_contained(c1, p):
    (x1, y1, z1) = c1
    (x, y, z) = p
    
    if (x1[0] <= x and x <= x1[1] and
        y1[0] <= y and y <= y1[1] and
        z1[0] <= z and z <= z1[1]):
        return True
    return False
    

    
print(is_fully_contained(input[0][1], input[1][1]))
print(is_conflict(input[0][1], input[1][1]))
```
